Importing Libraries

Quantifying Survival Percentage

```python
survival_percentage = (df['Survived'].sum() / len(df)) * 100

print(f"Percentage of people who survived: {survival_percentage:.2f}%")
```

Countplot for Survival Distribution (yes/no)

The port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
embarked_survival = pd.crosstab(df['Embarked'], df['Survived'], margins=True, margins_name='Total')

embarked_survival
```

Barplot for Survival Distribution by Port of Embarkation

```python
embarked_survival.drop('Total', axis=1, inplace=True) 
embarked_survival.plot(kind='bar', color=['#FF726F', '#74C9D2'], figsize=(10, 6))
plt.title('Survival Distribution by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

Calculate survival rate (percentage) for each combination of "Embarked" and "Survived"

```python
embarked_survival_rate = pd.crosstab(index=df['Embarked'], columns=df['Survived'], normalize='index') * 100
embarked_survival_rate
```

Transform Categorical Data into Numerical Data

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df['Sex']= labelencoder.fit_transform(df['Sex'])

df.head(10)
```

Split the Data and Initialize a Logistic Regression Model

```python
X= df[['Pclass', 'Sex']]
Y=df['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

Model Prediction

```python
pred = print(log.predict(X_test))
```

```python
print(Y_test)
```

Machine Learning Model for Prediction of an Outcome

```python
import warnings
warnings.filterwarnings("ignore")

res= log.predict([[0,1]])

if(res==0):
  print("So Sorry! Not Survived")
else:
  print("Survived")
```
